**Data Cleaning;Filling Missing Data**

```python
sns.heatmap(mhouse.isnull())
```

```python
mhouse['BuildingArea'].fillna(value=mhouse['BuildingArea'].mean(),inplace=True)
mhouse['YearBuilt'].fillna(value=mhouse['YearBuilt'].mean(),inplace=True)
```

```python
sns.heatmap(mhouse.isnull())
```

```python
mhouse.drop('CouncilArea',axis=1,inplace=True)
```

```python
mhouse.isnull().sum()
```

```python
mhouse['Car'].fillna(value=mhouse['Car'].mean(),inplace=True)
```

```python
mhouse.isnull().sum()
```

```python
mhouse.corr()['Price'].sort_values()
```

**Attaching Values to Y and X Variables**

```python
X = mhouse[['YearBuilt','Lattitude','Distance','Propertycount','Landsize','BuildingArea','Postcode','Longtitude',
    'Car','Bathroom','Bedroom2','Rooms']]
y = mhouse['Price']
```

```python
X.describe()
```

**Model Selection Using Scikit Learn**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=101)
```

**Importing Decision Tree Regressor Using Scikit Learn**

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor()
```

**Fitting The Data Into Our Model**

```python
dtr.fit(X_train,y_train)
```

**Testing Our Model Using The Test Data**

```python
pred = dtr.predict(X_test)
```

```python
pred
```

**Using Our Model To Predict Outcomes**

```python
print("Making predictions for the following 5 houses:")
print(X.tail(7))
print("The predictions are")
print(dtr.predict(X.tail(7)))
```

**Validating Our Data Using Metrics From SciKit Learn**

```python
from sklearn import metrics
```

```python
predp = pred
```

**The Mean Absolute Error and The Mean Squared Error Of Our Model**

```python
metrics.mean_absolute_error(y_test,predp)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
```
